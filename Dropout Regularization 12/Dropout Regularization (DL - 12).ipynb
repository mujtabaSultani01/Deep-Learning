## Dropout Regularization
**Dropout regularization** is a technique used in deep learning to prevent overfitting. It works by randomly 'dropping out' (i.e. setting to zero) a number of neurons during each training iteration. This helps to reduce the complexity of the model and make the learning more robust.

This dataset describes sonar chirp returns bouncing off different services. The 60 input variables are the strength of the returns at different angles. It's a binary classification problem that requires a model to differentiate rocks from metal cylinders.

* Dataset information: https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks) 
* Dataset Download link: https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data

```python
# Require libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# For warning ignorance
import warnings
warnings.filterwarnings('ignore')
```

```python
# Let's first read dataset:
df = pd.read_csv("sonar_dataset.csv")
df.head()
```

```python
# We ignore the header because the dataset doesn't has column names, so it will use integer sequence as column names:
df = pd.read_csv("sonar_dataset.csv", header=None)
df.head()
```

```python
# Let's see the columns names which is numeric integer from 0 to 60:
df.columns
```

* So here from 0 to 59 is our independent variables and the 60th one is our target variable.

```python
# Let's see the number of rows and columns in the dataset:
df.shape
```

```python
# To check n.a. values in the dataset:
df.isnull().sum()
```

```python
# To check the number of different values (number of classes) in the target variable:
df[60].value_counts()
```

* So we have 111 values of 'M' and 97 values of 'R'
* We know it's binary classification problem.

```python
# The next step is to create 'X' and 'Y'.
X = df.drop([60], axis = "columns")
y = df[60]
y.sample(6)
```

```python
# So as we see the 'y' is text data and we need to convert it into integer. So we just use get_dummies method to do One Hot Encodein and then we drop the first dummy:
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
# Again if we do value_count(), we will see the convert result:
y.value_counts()
```

```python
# The next step is to divide the data into train and test dataset:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
# To check the some random train samples:
X_train.sample(5)
```

```python
# And also some random test samples:
X_test.sample(5)
```

```python
# The next step is to create Artificial Neural Network (ANN), so for that we need to import tensorflow and keras:
import tensorflow as tf
from tensorflow import keras as kr
```

```python
# So for ANN model creation we call to sequential function and define the parameters:
model = kr.Sequential([
    kr.layers.Dense(60, input_dim=60, activation='relu'),
    kr.layers.Dense(30, activation='relu'),
    kr.layers.Dense(15, activation='relu'),
    kr.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy', 
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=100, batch_size=8) # here we use mini-batch gradient descent. after each 8 samples the error will be calculated and then back propagation is done.
```

* So after 80 epochs we got 100% accuracy.

```python
# So we want to overfit this model. Now the model did really good. let's see the performance on test dataset:
model.evaluate(X_test, y_test)
```

```python
# To see some prediction, how the model is good:

y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])
```

```python
# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
# Now to display y_test and compare it with the model prediction:
y_test[:10]
```

```python
# To see the classification report:
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test, y_pred))
```

```python
# So now let's try the model with a dropout layer:
modeld = kr.Sequential([
    kr.layers.Dense(60, input_dim=60, activation='relu'),
    kr.layers.Dropout(0.5), # It will drop the 50% of the neurons.
    kr.layers.Dense(30, activation='relu'),
    kr.layers.Dropout(0.5),
    kr.layers.Dense(15, activation='relu'),
    kr.layers.Dropout(0.5),
    kr.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
# Now we find 88% accuracy on train dataset. The previous one was 100%. We see that the dropout layer reduce the accuracy on
# train dataset, but it's ok. We take care most about test set. The previous test set score were 77%. let's see how many is
# it here right now?
modeld.evaluate(X_test, y_test)
```

* Now as we can see that by using dropout layer we got better performance.

```python
# Now again if we compare the results:
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
# To print the classification report from the new model:
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test, y_pred))
```

* We see the improvements in classification report.
