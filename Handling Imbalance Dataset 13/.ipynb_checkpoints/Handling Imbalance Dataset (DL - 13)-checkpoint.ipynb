# Handling Imbalance Dataset
**An imbalanced dataset** is a dataset where the target classes are not equally represented. This can be an issue for machine learning models, as they can become biased towards the majority class and fail to accurately identify the minority class. To address this issue, data scientists often use resampling techniques to balance the dataset.
* There are several methods that can be used to handle imbalanced datasets in deep learning, including data augmentation, oversampling, undersampling, and class weighting. Each of these methods has its own advantages and disadvantages, so it's important to evaluate each one carefully to determine which is best suited for the task at hand.

* The first technique to handle imbalance dataset is **Under sampling majority class.** Let's say we have 99000 samples belonging to one class and 1000 samples belonging to another class. Let's say this is fraud detection scenario where 1000 transactions are fraud and 99000 transactions are not fraud. So to tackle this imbalance we take randomly 1000 samples from 99000 samples class and we discard remaining samples and then we combine these 1000 samples with the red class 1000 samples and we train our model. Obviously this is not the best approach because we throwing a way so much data.

<img src = "img.jpg" width = "500px" height = "400px"></img>

* The 2nd technique is **Over sampaling manority class by duplication.** Here in this technique we duplicate the minority class 99 times. So as result we'll get 99000 samples and then we train the machine learning model.

<img src = "img1.jpg" width = "500px" height = "400px"></img>

* The 3rd technique is **Over sampaling minority class using SMOTE.** So here we use K-Nearest Neighbor (KNN) Algorithm and we try to produce synthetic samples from the 1000 samples. In python there is a model called **imbalance-learn** which can be use for smothing.

<img src = "img2.jpg" width = "500px" height = "400px"></img>

* The 4th technique is **Ensemble.** Let's see you have 3000 transactions in one class and 1000 transactions in other class. Now you divide those 3000 class into three parts. 1st you take the first part and you combine it with the 1000 red samples and you train the model. 2nd you take the second part and you combine it with the red samples and you train the model. 3rd you take the third part and combine it with the red samples and train the model. The as result you will have three models and then you use a majority vote. Simple you may use Random Forest and will take the majority vote.

<img src = "img3.jpg" width = "500px" height = "400px"></img>

* The 5th approach is **Focal Loss.** Focal Loss will penalize majority samples during loss calculations and give more weights to minority class samples.

### Building a ANN model for handling imbalance dataset

```python
# So simple changes is made in prvious notebook, for the model a separate function is created.
# The weights parameter is also added, which is normally using for tackling imbalance problem.
def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model = keras.Sequential([
        keras.layers.Dense(26, input_dim=26, activation='relu'),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
    
    if weights == -1:
        model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight = weights)
    
    print(model.evaluate(X_test, y_test))
    
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    
    print("Classification Report: \n", classification_report(y_test, y_preds))
    
    return y_preds
```

```python
# Call for the function
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

* So in upper result we see that precision, reall and F1-score for class 1 is very low. Our main goal is to balance the classes and bring the class1 F1-score higher.
* Class 1 is the people who leaving the business.
* Generally the goals is to improve the prediction accuracy for both classes equally.
* So the first technique which we try is under samplaing.

### Method 1: Under Sampaling Technique

```python
# Count the number of samples in each class.
count_class_0, count_class_1 = df1.Churn.value_counts()

# Divide by class
# We just store 0 and 1 samples in different DataFrames.
df_class_0 = df2[df2['Churn'] == 0]
df_class_1 = df2[df2['Churn'] == 1]
```

```python
count_class_0, count_class_1
```

```python
# Now if we shape the df_classes, we will see the imbalance of data,
# df_class_0
df_class_0.shape
```

```python
df_class_1.shape
```

```python
# So now we'll under sample the df_class_0.
# So to do under sampling, first sample() is a function in DataFrame which return the given number samples from a DataFrame randomly.
# So we want count_class_1 number of samples from df_class_0. And we do this using sample() function.
df_class_0.sample(count_class_1).shape # When we shape it, it will create similar samples with df_class_1.
```

```python
# So we'll store the samples in a variable.
df_class_0_under = df_class_0.sample(count_class_1)
```

```python
# So the next thing is to combine both DataFrame together.
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)
df_test_under.shape

print('Random under-sampling:')
print(df_test_under.Churn.value_counts())
```

```python
# So now we create X and Y from this new dataset and call for the train_test _split method.
X = df_test_under.drop('Churn',axis='columns')
y = df_test_under['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y) # Stratify argument just show that you have balance dataset.
```

```python
# Let's see the balance of y_train samples:
y_train.value_counts()   # For both classes will be similar because of stratify argument.
```

```python
# Now using the new samples we train the model:
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

* Now if we compare the precision, recall and F1-scores for the class1, we'll see the improvements.

### Method 2: Oversampaling Technique

```python
# So let's again print the classes count to see the imbalance of data:
count_class_0, count_class_1 = df1.Churn.value_counts()
count_class_0, count_class_1
```

```python
# So the count_class_1 Dataset we make oversample. So for example if I want to create 2000 samples from this class, I will do like this:
df_class_1.sample(200, replace = True).shape
```

```python
# Similarly we need 'count_class_0' number oversampling, so we put count_class_0 instead of 2000:
df_class_1.sample(count_class_0, replace = True).shape
```

```python
# So now we have 5163 samples in class 1 is well. 
# So we want to store this samples in a variable. Additionaly we combine both of them using concat method.
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_test_over = pd.concat([df_class_0, df_class_1_over], axis=0)

print('Random over-sampling:')
print(df_test_over.Churn.value_counts())
```

```python
# Next we create X and Y. And also we call train_test_split method.
X = df_test_over.drop('Churn',axis='columns')
y = df_test_over['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

### Method 3: SMOTE

```python
# First we get X and y from the original df2 DataFrame:
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
# To check the imbalance:
y.value_counts()
```

```python
# Next we import SMOTE.
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority') # The way we create SMOTE.
X_sm, y_sm = smote.fit_resample(X, y)   # Create new samples X_sm, y_sm

y_sm.value_counts()
```

```python
# The next step is to call train_test_split method:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
y_train.value_counts()
```

```python
# Next we train the model:
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

### Method 4: Ensemble

```python
# So again we take the original Dataset, and check the imbalance:
df2.Churn.value_counts()
```

```python
# Next we create X and y:
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
# Then we create X train and Y train:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
# Check the imbalance:
y_train.value_counts()
```

```python
# Now let's see the ratio between two classes
4130/1495
```

```python
# ...
df3 = X_train.copy()
df3['Churn'] = y_train
```

```python
# We see the ratio which is almost 3, so we divid the class 0 into three parts. 
# So first let's create new DataFrames for class 0 and class 1 samples:
df3_class0 = df3[df3.Churn==0]
df3_class1 = df3[df3.Churn==1]
```

```python
# The next stpe is to split the class 0 into three parts and create X_train and y_train using the following function:
def get_train_batch(df_majority, df_minority, start, end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Churn', axis='columns')
    y_train = df_train.Churn
    return X_train, y_train  
```

```python
# Call to the function:
X_train, y_train = get_train_batch(df3_class0, df3_class1, 0, 1495)
```

```python
# Next is again to call the model and store the result as model1, because we create want to create three different model.
y_pred1 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
# The second model is:
X_train, y_train = get_train_batch(df3_class0, df3_class1, 1495, 2990)

y_pred2 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
# The third model is:
X_train, y_train = get_train_batch(df3_class0, df3_class1, 2990, 4130)

y_pred3 = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
# So now we have three different votes and we want to take the majority vote.
# The majority vote is assessing as follow:

y_pred_final = y_pred1.copy()
for i in range(len(y_pred1)):
    n_ones = y_pred1[i] + y_pred2[i] + y_pred3[i]
    if n_ones>1:
        y_pred_final[i] = 1
    else:
        y_pred_final[i] = 0
```

```python
# To print the classification report for this ensemble approach:
cl_rep = classification_report(y_test, y_pred_final)
print(cl_rep)
```
